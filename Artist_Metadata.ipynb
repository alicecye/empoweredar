```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from pathlib import Path
import requests
import sys
```

```python
cid = Path('_keys/clientId.txt').read_text()
secret = Path('_keys/secret.txt').read_text()

client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
monolink_uri = 'spotify:artist:2I4hRNCYkPKJQlkoEZKjYx'
```

```python
#make call to api with monolink uri to fetch albums 
album_results = sp.artist_albums(monolink_uri, album_type='album')

#grab albums from results
albums = album_results['items']

#return paged results 
while album_results['next']:
    album_results = spotify.next(album_results)
    albums.extend(album_results['items'])
```

```python
for album in albums:
    print(album['name'])
```

```python
#Artist Catalog
artist_catalog = pd.DataFrame(columns=['artists','album_uri','id','name','album_urls','spotify_album_url','release_date','total_tracks','available_markets','num_avail_markets'])

for i in range(len(albums)):
    artist_catalog = artist_catalog.append({'artists':albums[i]['artists'],
                                           'album_uri':albums[i]['uri'],
                                           'id':albums[i]['id'],
                                           'name':albums[i]['name'],
                                           'album_urls':albums[i]['external_urls'],
                                           'spotify_album_url':albums[i]['external_urls']['spotify'],
                                           'release_date':albums[i]['release_date'],
                                           'total_tracks':albums[i]['total_tracks'],
                                           'available_markets':albums[i]['available_markets'],
                                           'num_avail_markets': len(albums[i]['available_markets'])},ignore_index=True)
    i += 1
```

```python
for i in range(data.shape[0]):
    uri = data['uri'][i]
    print(uri)
    
    artist_output = sp.artist(uri)

    artist_meta = pd.DataFrame(columns=['artist_url','uri','artist_name','artist_popularity','artist_image','genres','followers'])

    artist_url = artist_output['external_urls']['spotify']
    artist_uri = artist_output['uri']

    artist_meta = artist_meta.append({'artist_url':artist_url,
                                     'uri':artist_uri,
                                     'artist_name':artist_output['name'],
                                     'artist_popularity':artist_output['popularity'],
                                     'artist_image':artist_output['images'][0]['url'],
                                     'genres':artist_output['genres'],
                                     'followers':artist_output['followers']['total']}, ignore_index=True)
    
    related = sp.artist_related_artists(uri)
    #artist_meta['related_artists'] = related.copy()
    
    related_pop_total = 0

    for artist in related['artists']:
        related_pop_total += artist['popularity']

    if related_pop_total == 0:
        artist_meta['avg_related_artist_pop'] = 'None'
    else:
        artist_meta['avg_related_artist_pop'] = related_pop_total/len(related['artists'])
    
    data = data.merge(artist_meta, how='left', on='uri')
```
