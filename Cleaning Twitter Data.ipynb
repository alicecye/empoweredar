```python
import pandas as pd
import datetime
```

```python
data=pd.read_csv('twitter_data (1).csv')
```

```python
data.head()
```

```python
data.columns
```

```python
cols = ['created_at', 'id_str', 'retweet_count', 'favorite_count', 'favorited', 
        'retweeted', 'lang', 'user.id_str', 'user.name', 'user.screen_name', 
        'user.followers_count', 'user.friends_count', 'user.created_at', 'user.favourites_count', 
        'user.statuses_count', 'user.description', 'text']
```

```python
data_short = data[cols]
```

```python
data_short.dtypes
```

```python
# drop time components and reformat date components
data_short['created_at'] = pd.to_datetime(data_short['created_at']).dt.to_period("D")
data_short['user.created_at'] = pd.to_datetime(data_short['user.created_at']).dt.to_period("D")
```

```python
pd.options.display.max_colwidth = 500
```

```python
# view sample tweet with messy text
print(data_short.loc[data_short['id_str'].isin(['1541499401516072963']), 'text']
     )
```

```python
# clean up all fields by removing newline, tabs, etc.
data_short.replace(to_replace=[r"\\t|\\n|\\r", "\t|\n|\r"], value=[" "," "], regex=True, inplace=True)
data_short.replace(to_replace=[r"  "], value=[" "], regex=True, inplace=True)
```

```python
data_short.to_csv("twitter_subset.tsv", sep="\t", index=False)
```
