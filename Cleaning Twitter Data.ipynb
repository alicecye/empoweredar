```python
data_short = pd.read_csv('twitter_subset.tsv', sep='\t')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
import re
```

```python
nltk.download(["names","stopwords","state_union","twitter_samples","movie_reviews", "averaged_perceptron_tagger","vader_lexicon","punkt"])
sia = SentimentIntensityAnalyzer()
```

```python
def get_sentiment(df):
  empty = []

  for tweet in df.text:

    if pd.isna(tweet):
      tweet = ''

    ## clean tweets
    
    ### remove any twitter user names
    tweet = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet).split())
    ### remove hashtags
    tweet = ' '.join(re.sub("(#[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet).split())

    output = sia.polarity_scores(tweet.lower())
    empty.append(output)
  return empty

sentiment_scores = pd.DataFrame(get_sentiment(data_short))
final_sentiment = pd.concat([data_short, sentiment_scores], axis=1)
```
